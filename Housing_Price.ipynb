```python
# Simple Multiple Linear Regression Model
features =["floors", "waterfront","lat" ,"bedrooms" ,"sqft_basement" ,"view" ,"bathrooms","sqft_living15","sqft_above","grade","sqft_living"]
M=df[features]
Y = df['price']
lm = LinearRegression()
lm.fit(M,Y)
lm.score(M, Y)  #gives R² on training data only.
```

```python

# ---------------------------------------------------------------
# ▶️ Model 2: Pipeline with Scaling + PolynomialFeatures
# ---------------------------------------------------------------
# This model uses a pipeline with three steps:
#     1. StandardScaler()         → Scales features (mean=0, std=1)
#     2. PolynomialFeatures()     → Generates higher-order and interaction terms
#     3. LinearRegression()       → Fits linear model on these expanded features
#
# Effectively, this becomes a Polynomial Regression model
# which can capture non-linear relationships between features and target.
#
# Example:
#     Input = [
#         ('scale', StandardScaler()),
#         ('polynomial', PolynomialFeatures(include_bias=False)),
#         ('model', LinearRegression())
#     ]
#     pipe = Pipeline(Input)
#     pipe.fit(M, Y)
#     ypipe = pipe.predict(M)
#     print(r2_score(Y, ypipe))   # R² for polynomial (non-linear) model
#
# ---------------------------------------------------------------
# 🧠 Summary:
# Model 1 → Simple Linear, no scaling, less flexible
# Model 2 → Polynomial (non-linear), scaled data, more flexible
#
# ⚠️ Note: Polynomial models can easily overfit if degree is too high.
# ---------------------------------------------------------------

#Include bias = False, means no intercept terms
```

```python
from sklearn.metrics import r2_score
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler,PolynomialFeatures
```

```python
#Now better way, Polynomial model
Input=[
    ('scale', StandardScaler()),
    ('polynomial', PolynomialFeatures(include_bias=False)),
    ('model', LinearRegression())]
```

```python
pipe = Pipeline(Input)
pipe.fit(M,Y)
ypipe = pipe.predict(M)
print(r2_score (Y, ypipe))
```

```python
# ---------------------------------------------------------------
# 💡 Why Polynomial Regression often gives higher R² than Linear:
# ---------------------------------------------------------------
# 1. More flexible — can model nonlinear curves in data.
# 2. Includes interaction terms → captures combined effects of features.
# 3. Adding polynomial terms increases predictors, and R² on training data
#    can only improve or stay the same (never decrease).

# Higher R² on training data doesn’t always mean a better model.
# Polynomial models can easily overfit — always check R² on test data.
```

```python
#------------------Always Best to Split Data b/w train and test-----------------------------

# If Train R² ≫ Test R², your model overfits (too complex).
# If both are reasonable and close → good predictive performance.

# random_state locks the randomness.
# Same random_state → same train/test split every time (reproducible results).
# No random_state → different split each run.
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, PolynomialFeatures
from sklearn.pipeline import Pipeline
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

features = ["floors", "waterfront", "lat", "bedrooms", "sqft_basement",
            "view", "bathrooms", "sqft_living15", "sqft_above", "grade", "sqft_living"]

X = df[features]
Y = df['price']

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
lm.fit(X_train, Y_train)

print("Train R²:", lm.score(X_train, Y_train))
print("Test R² :", lm.score(X_test, Y_test))
```

```python

features = ["floors", "waterfront", "lat", "bedrooms", "sqft_basement",
            "view", "bathrooms", "sqft_living15", "sqft_above", "grade", "sqft_living"]

M = df[features]
Y = df['price']

M_train, M_test, Y_train, Y_test = train_test_split(M, Y, test_size=0.2, random_state=42)

pipe = Pipeline([
    ('scale', StandardScaler()),
    ('poly', PolynomialFeatures(degree=2, include_bias=False)),
    ('model', LinearRegression())
])

pipe.fit(M_train, Y_train)

print("Train R²:", pipe.score(M_train, Y_train))
print("Test R² :", pipe.score(M_test, Y_test))
```

```python
#  Improved prediction accuracy with ~8% lower RMSE compared to baseline model.
# lower RMSE better it is
from sklearn.metrics import mean_squared_error
import numpy as np

# Model 1
y_pred1 = lm.predict(X_test)
rmse1 = np.sqrt(mean_squared_error(Y_test, y_pred1))
print("Model 1 RMSE:", rmse1)

# Model 2
y_pred2 = pipe.predict(M_test)
rmse2 = np.sqrt(mean_squared_error(Y_test, y_pred2))
print("Model 2 RMSE:", rmse2)

improvement = ((rmse1 - rmse2) / rmse1) * 100
print(f"Improvement in RMSE: {improvement:.2f}%")
```

```python
# Model Evaluation and Refinement
# regularized regression, which is essential for handling multicollinearity and overfitting
# Ridge Regression adds a penalty (alpha) to control large coefficients.
# alpha = 0 → same as LinearRegression
# Higher alpha → more regularization (simpler, less overfitting)
# Useful when features are correlated or model overfits the training data.
```

```python
from sklearn.linear_model import Ridge
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score

M_train, M_test, Y_train, Y_test = train_test_split(M, Y, test_size=0.2, random_state=42)

ridge = Ridge(alpha=0.1)   # α = 0.1 means mild regularization
ridge.fit(M_train, Y_train)

print("Train R²:", ridge.score(M_train, Y_train))
print("Test R² :", ridge.score(M_test, Y_test))
```

```python
# COMPARISON
from sklearn.linear_model import LinearRegression, Ridge

lm = LinearRegression().fit(M_train, Y_train)
ridge = Ridge(alpha=0.1).fit(M_train, Y_train)

print("Linear Regression R² (Train):", lm.score(M_train, Y_train))
print("Ridge Regression R² (Train):", ridge.score(M_train, Y_train))


print("Linear Regression R² (Test) :", lm.score(M_test, Y_test))
print("Ridge Regression R² (Test) :", ridge.score(M_test, Y_test))
```

```python
# ABOVE there is no much significance difference probably because no Overfitting, No Multicollinearity
# Linear Regression → higher Train R² but lower Test R² (overfitting)
# Ridge Regression → slightly lower Train R² but higher Test R² (better generalization)
```

```python
# GRPAHICALLY
# (R² vs α)
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import Ridge

alphas = np.logspace(-3, 3, 20)  # alpha values from 0.001 to 1000
train_scores = []
test_scores = []

for a in alphas:
    ridge = Ridge(alpha=a)
    ridge.fit(M_train, Y_train)
    train_scores.append(ridge.score(M_train, Y_train))
    test_scores.append(ridge.score(M_test, Y_test))

plt.figure(figsize=(7, 5))
plt.plot(alphas, train_scores, label="Train R²", marker='o')
plt.plot(alphas, test_scores, label="Test R²", marker='s')
plt.xscale('log')
plt.xlabel("Alpha (Regularization Strength)")
plt.ylabel("R² Score")
plt.title("Effect of Ridge Regularization (R² vs α)")
plt.legend()
plt.grid(True)
plt.show()

```
